```python
# How to rename the columns:
df=pd.DataFrame({"a":[1,2,3,4],'b':[5,6,7,8]})
df
```

```python
df.rename(columns={'a':'col_a','b':'col_b'},inplace=True)
df
```

```python
df.columns=['col_aa','csj_s']
df
```

```python
#  How to replace ny character ,string:
df.columns=df.columns.str.replace('_',"* ")
df
```

```python
# How to add prefix 
df=df.add_prefix('baba_')
df
```

```python
# How to add suffix:
df=df.add_suffix('baba_')
df
```
